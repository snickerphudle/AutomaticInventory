```
img_path = '/content/drive/MyDrive/FutureMakers_Code/FutureMakers_custom_YOLO_2021/custom_YOLO/images/'
import os
import subprocess
os.listdir(img_path)
files = os.listdir(img_path)
fp=open('results.csv', 'w')
for file in files:
  print (file)
  if file.endswith('.jpg'):
    result = subprocess.run(['./darknet', 'detector', 'test', 'YOLO_labeled_groceries/groceries_all.data', 'groceries_yolov4.cfg', 'backup/groceries_yolov4_best.weights', '/content/drive/MyDrive/FutureMakers_Code/FutureMakers_custom_YOLO_2021/custom_YOLO/images/' + file], stdout=subprocess.PIPE)
    #print (result.stdout)
    print(str(result).split('seconds')[0].split('.jpg')[0].split('/')[-2], str(result).split('seconds')[1].replace('\\n',';'))
    fp.write(str(result).split('seconds')[0].split('.jpg')[0].split('/')[-2] + ',' + str(result).split('seconds')[1].replace('\\n',';'))
    #print(str(result).split('%')[0])
    #fp.write(str(result).split('seconds')[1])
    #break
    fp.write('\n')
fp.close()
```
